##**Regresión Lineal**

```python
#Grafica de dispersión salario Vs experiencia
x=salarios["Añosdeexperiencia"]
y=salarios["Salario"]

#scater es un tipo de grafica de dispersion el primer valor es el dato x y el segundo valor es y para poder generar la grafica
plt.scatter(x,y,color="red")

#Titulo del grafico
plt.title("Salario vs Años de experiencia")

#titulo eje x
plt.xlabel("Añosdeexperiencia")

#titulo eje y
plt.ylabel("Salario")

plt.show()
```
